# KARA LangChain Integration

## 1. Setup

```python
from langchain_core.documents import Document
from kara.integrations.langchain import KARATextSplitter
```

## 2. Initialization

```python
# Initialize KARA splitter with precise OpenAI token counting
splitter = KARATextSplitter.from_tiktoken_encoder(
    encoding_name="cl100k_base",
    chunk_size=10,
    chunk_overlap=2
)
```

## 3. Initial Document Processing

```python
text = """
The Knowledge-Aware Re-embedding Algorithm (KARA) is designed to optimize RAG workflows. 
By identifying unchanged segments of text, it prevents redundant embedding operations.
This is particularly useful when documentation is updated frequently.
Standard splitters would re-calculate every chunk even for a small change.
KARA, however, maintains a graph of units to find the most efficient path.
"""

doc = Document(page_content=text, metadata={"source": "kara_docs.txt"})
initial_chunks = splitter.split_documents([doc])

print(f"Initial chunks: {len(initial_chunks)}")
for i, chunk in enumerate(initial_chunks):
    print(f"Chunk {i}: {chunk.page_content[:60]}...")
```

## 4. Efficient Document Update

```python
# Realistic update: inserting content in the middle
updated_text = """
The Knowledge-Aware Re-embedding Algorithm (KARA) is designed to optimize RAG workflows. 
By identifying unchanged segments of text, it prevents redundant embedding operations.
KARA supports various tokenizers, including OpenAI (via tiktoken) and Hugging Face.
This is particularly useful when documentation is updated frequently.
Standard splitters would re-calculate every chunk even for a small change.
KARA, however, maintains a graph of units to find the most efficient path.
"""

updated_doc = Document(page_content=updated_text, metadata={"source": "kara_docs.txt"})
updated_chunks = splitter.split_documents([updated_doc])

print(f"Updated chunks: {len(updated_chunks)}")
```

## 5. Efficiency Metrics

```python
res = splitter.last_result
print(f"Efficiency Summary:")
print(f"- Reused Chunks: {res.num_reused}")
print(f"- New Chunks: {res.num_added}")
print(f"- Efficiency Ratio: {res.efficiency_ratio:.2%}")
```
